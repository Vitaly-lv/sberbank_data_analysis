Для календаря 2021 года, определим праздничные дни, так как они не все указаны в загруженном файле

```python
holidays = [pd.to_datetime('01.01.2021'),
            pd.to_datetime('01.04.2021'),
            pd.to_datetime('01.05.2021'),
            pd.to_datetime('01.06.2021'),
            pd.to_datetime('01.07.2021'),
            pd.to_datetime('01.08.2021'),
            pd.to_datetime('02.22.2021'),
            pd.to_datetime('02.23.2021'),
            pd.to_datetime('03.08.2021'),
            pd.to_datetime('05.03.2021'),
            pd.to_datetime('05.10.2021'),
            pd.to_datetime('06.14.2021'),
            pd.to_datetime('11.04.2021'),
            pd.to_datetime('11.05.2021'),
            pd.to_datetime('12.31.2021')]
```

```python
# кастомизация календаря на 2021 год
from pandas.tseries.holiday import AbstractHolidayCalendar, Holiday,\
nearest_workday, EasterMonday
from pandas.tseries.offsets import DateOffset, Day

class RussianBusinessCalendar(AbstractHolidayCalendar):
    start_date = datetime.datetime(2021, 1, 1)
    end_date = datetime.datetime(2021, 12, 31)
    
workday_2021 = pd.offsets.CustomBusinessDay(weekmask='Mon Tue Wed Thu Fri', 
                                              calendar=RussianBusinessCalendar(),
                                              holidays = holidays)
```

```python
#Вставляем столбец 'выходной' с данными
columns = ['Выходной']
dti = pd.date_range(start='2021-01-01', end='2021-12-31', freq='D')
calendar2021 = pd.DataFrame(index=dti, columns=columns)
wdti = pd.bdate_range('2021-01-01', '2021-12-31', freq='C',
                      weekmask='Mon Tue Wed Thu Fri', holidays=holidays) # Рабочие дни в 2021
for day in calendar2021.index:
    calendar2021['Выходной'][day] = int(day not in wdti)
```

```python
#Вставляем столбец 'Праздник' с данными
calendar2021['Праздник'] = np.nan
for day in holidays:
    calendar2021['Праздник'][day] = 1
calendar2021['Праздник'].fillna(0, inplace = True)
calendar2021.astype({"Праздник": int})
```

```python
#Вставляем столбец 'Предпраздничный рабочий день' с данными
pre_holiday_work_day =[pd.to_datetime('04.30.2021'),
                       pd.to_datetime('06.11.2021'),
                       pd.to_datetime('11.03.2021')]
calendar2021['Предпраздничный рабочий день'] = np.nan
for day in pre_holiday_work_day:
    calendar2021['Предпраздничный рабочий день'][day] = 1
calendar2021['Предпраздничный рабочий день'].fillna(0, inplace = True)
calendar2021.astype({'Предпраздничный рабочий день': int})
```

```python
#Вставляем столбец 'последний день месяца' с данными
calendar2021['Последний день месяца'] = np.nan
for day in calendar2021.index:
    calendar2021['Последний день месяца'][day] = int(day.is_month_end)
calendar2021.astype({'Последний день месяца': int}) 
```

```python
#Вставляем столбец 'Последний рабочий день месяца' с данными
last_month_day = []
for day in calendar2021.index:
    if day.is_month_end:
        last_month_day.append(day) # Список последних дней месяца

last_month_day = pd.to_datetime(last_month_day)
last_work_month_day = last_month_day - workday_2021 # переместить на последний рабочий день
```

```python
calendar2021['Последний рабочий день месяца'] = np.nan
for day in last_work_month_day:
    calendar2021['Последний рабочий день месяца'][day] = 1
calendar2021['Последний рабочий день месяца'].fillna(0, inplace = True)
calendar2021.astype({'Последний рабочий день месяца': int}) 
```

```python
#Вставляем столбцы дни недели с данными
weekday = calendar2021.index.day_name()
week_day = pd.get_dummies(weekday)
calendar2021['Пн'] = week_day['Monday'].values
calendar2021['Вт'] = week_day['Tuesday'].values
calendar2021['Ср'] = week_day['Wednesday'].values
calendar2021['Чт'] = week_day['Thursday'].values
calendar2021['Пт'] = week_day['Friday'].values
calendar2021['Сб'] = week_day['Saturday'].values
calendar2021['Вс'] = week_day['Sunday'].values
```

```python
#Вставляем столбцы месяцы с данными
month = calendar2021.index.month_name()
month_name = pd.get_dummies(month)
calendar2021['Январь'] = month_name['January'].values
calendar2021['Февраль'] = month_name['February'].values
calendar2021['Март'] = month_name['March'].values
calendar2021['Апрель'] = month_name['April'].values
calendar2021['Май'] = month_name['May'].values
calendar2021['Июнь'] = month_name['June'].values
calendar2021['Июль'] = month_name['July'].values
calendar2021['Август'] = month_name['August'].values
calendar2021['Сентябрь'] = month_name['September'].values
calendar2021['Октябрь'] = month_name['October'].values
calendar2021['Ноябрь'] = month_name['November'].values
calendar2021['Декабрь'] = month_name['December'].values
```

```python
#Вставляем столбец 'Предновогодний выходной день'  с данными
calendar2021['Предновогодний выходной день'] = np.nan
calendar2021['Предновогодний выходной день']['31.12.2021'] = 1
calendar2021['Предновогодний выходной день'].fillna(0, inplace = True)
calendar2021.astype({'Предновогодний выходной день': int})  
```

```python
ts = pd.read_csv('sber_stocks.csv', usecols=['CLOSE', 'TRADEDATE'], 
                 index_col='TRADEDATE', parse_dates=True)
ts = ts.CLOSE
```

```python
df_ts = pd.DataFrame(ts.asfreq('1D').fillna(method='ffill'))
df_ts['weekday'] = df_ts.index.weekday
df_ts.head()
```

```python
from pandas.tseries.holiday import AbstractHolidayCalendar, Holiday, nearest_workday, EasterMonday
from pandas.tseries.offsets import DateOffset, Day

class RussianBusinessCalendar(AbstractHolidayCalendar):
    start_date = datetime.datetime(2013, 3, 25)
    end_date = datetime.datetime(2020, 1, 10)
    rules = [Holiday('Пр1', month = 1, day = 1), 
             Holiday('Пр2', month = 1, day = 2), 
             Holiday('Пр3', month = 1, day = 3), 
             Holiday('Пр4', month = 1, day = 4), 
             Holiday('Пр5', month = 1, day = 5), 
             Holiday('Пр6', month = 1, day = 6), 
             Holiday('Пр7', month = 1, day = 7), 
             Holiday('Пр8', month = 1, day = 8), 
             Holiday('Пр9', month = 2, day = 23), 
             Holiday('Пр10', month = 3, day = 8), 
             Holiday('Пр11', month = 5, day = 1), 
             Holiday('Пр12', month = 5, day = 9), 
             Holiday('Пр13', month = 6, day = 12), 
             Holiday('Пр14', month = 11, day = 4)]
russian_busday = pd.offsets.CustomBusinessDay(weekmask='Mon Tue Wed Thu Fri', 
                                              calendar=RussianBusinessCalendar())
```

Лаговые переменные рабочих дней с глубинами: 1, 3, 5

```python
sh1 = ts.asfreq('1D').fillna(method='ffill').shift(1, freq=russian_busday)
sh3 = ts.asfreq('1D').fillna(method='ffill').shift(3, freq=russian_busday)
sh5 = ts.asfreq('1D').fillna(method='ffill').shift(5, freq=russian_busday)
```

```python
final = df_ts.merge(sh1, how = 'left', left_index=True, right_index = True)
final = final.merge(sh3, how = 'left', left_index=True, right_index = True)
final = final.merge(sh5, how = 'left', left_index=True, right_index = True)
final.columns = ['CLOSE', 'Weekday', 'Shift1', 'Shift3', 'Shift5']
final.head()
```
